```python
#  merge - Merging DataFrames 
# 1. Merge two DataFrames on a common column Product_ID. 
df = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104],
    'Product_Name': ['Laptop', 'Phone', 'Tablet', 'Monitor'],
    'Price': [1000, 500, 300, 200]
})

df = pd.DataFrame({
    'Product_ID': [102, 103, 104, 105],
    'Category': ['Electronics', 'Electronics', 'Accessories', 'Gaming'],
    'Stock': [50, 30, 20, 10]
})

pd.merge(df, df, on='Product_ID')
```

```python
# 2. Perform a left join between two DataFrames based on Employee_ID. 
df_emp1 = pd.DataFrame({
    'Employee_ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing']
})

df_emp2 = pd.DataFrame({
    'Employee_ID': [2, 3, 5, 6],
    'Salary': [60000, 70000, 80000, 90000]
})

pd.merge(df_emp1, df_emp2, on='Employee_ID', how='left')
```

```python
# 3. Merge two DataFrames, keeping all rows from the left DataFrame and matching rows 
# from the right. 
pd.merge(df, df, on='Product_ID', how='left')
```

```python
# 4. Merge two DataFrames on multiple columns (e.g., City and Age). 
data1 = {
    'Name': ['John', 'Alice', 'Bob', 'Carol'],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago'],
    'Age': [25, 30, 25, 35]
}

data2 = {
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Lawyer'],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago'],
    'Age': [25, 30, 25, 35]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

merged_df = pd.merge(df1, df2, on=['City', 'Age'])
print(merged_df)
```

```python
# 5. Perform an inner join on two DataFrames with a common column Country. 
data1 = {
    'Name': ['John', 'Alice', 'Bob', 'Carol'],
    'Country': ['USA', 'Canada', 'USA', 'Mexico'],
    'Age': [25, 30, 25, 35]
}

data2 = {
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Lawyer'],
    'Country': ['USA', 'Canada', 'Germany', 'Mexico'],
    'Salary': [50000, 60000, 70000, 80000]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

merged_df = pd.merge(df1, df2, on='Country', how='inner')
print(merged_df)
```

```python
#6. Merge two DataFrames and keep all rows, even if there’s no match in the second DataFrame. 
data1 = {
    'Name': ['John', 'Alice', 'Bob', 'Carol'],
    'Country': ['USA', 'Canada', 'USA', 'Mexico'],
    'Age': [25, 30, 25, 35]
}

data2 = {
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Lawyer'],
    'Country': ['USA', 'Canada', 'Germany', 'Mexico'],
    'Salary': [50000, 60000, 70000, 80000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

merged_df = pd.merge(df1, df2, on='Country', how='outer')

print(merged_df)
```
